```python
import pandas
import seaborn as sns
import matplotlib as plt

pandas.options.display.max_columns = None
pandas.options.display.max_rows = None

videogame_sales_data = pandas.read_csv("../best_selling_games.csv")
sales_data = videogame_sales_data.iloc[:,2]

count = 0
for num in sales_data:
    sales_data[count] = int(num.replace(",", ""))
    count += 1

videogame_sales_data
```

```python
"1 | Which video game company has the most sales combined from all their video games?"
"2 | What is a noteable attribute of those in the top 20 in sales have in common?"
"3 | Which company has produced the most video games?"
"4 | Which company has made the least amount of sales?"
"5 | Which company has made the least amount of games?"
"6 | Is being muliplatform compatible that impactful in terms of sales?"
"7 | How many games are stuck to only one platform?"
"8 | How many games are multiplatform?"
"9 | Which company is least likely to make the next big game?"
"10 | Which company would be most likely to make the next big game?"
```

```python
'''1.) Which video game company has the most sales combined from all their video games?
'''

sales_dict = {}
company = videogame_sales_data.iloc[:,6]

for co in company:
    sales_dict[co] = videogame_sales_data.loc[videogame_sales_data["Publisher(s)"] == co]["Sales"].sum()

sales_dict

# Nintendo has the most sales
```

```python
"2.) What is a noteable attribute of those in the top 20 in sales have in common?"

top_20 = videogame_sales_data.head(20)

top_20

# Most are Multiplatform and are mostly produced by Nintendo
```

```python
"3.) Which company has produced the most video games?"

company = videogame_sales_data.iloc[:,6]
nintendo = company == 'Nintendo'
mojang = company == 'Mojang Studios'
rockstar = company == 'Rockstar Games'
ea = company == 'Electronic Arts'
pubg = company == 'PUBG Corporation'
namco = company == 'Namco'
relogic = company == 'Re-Logic / 505 Games'
bethesda = company == 'Bethesda Softworks'
blizzard = company == 'Blizzard Entertainment'
cd_projekt_red = company == 'CD Projekt'
xbox = company == 'Xbox Game Studios'
sega = company == 'Sega'
twok = company == '2K Games'
psygnosis = company == 'Psygnosis'
sony = company == 'Sony Computer Entertainment'
nintendo_pokemon = company == 'Nintendo / The Pokémon Company'

nintendo.sum(), mojang.sum(), rockstar.sum(), ea.sum(), pubg.sum(), namco.sum(), relogic.sum(), bethesda.sum(), blizzard.sum(), cd_projekt_red.sum(), xbox.sum(), sega.sum(), twok.sum(), psygnosis.sum(), sony.sum(), nintendo_pokemon.sum()

# Nintendo has produced the most games
```

```python
"4.) Which company has made the least amount of sales?"

for co in company:
    sales_dict[co] = videogame_sales_data.loc[videogame_sales_data["Publisher(s)"] == co]["Sales"].sum()

sales_dict

# Psygnosis has made the least amount of sales
```

```python
"5.) Which company has made the least amount of games?"

company = videogame_sales_data.iloc[:,6]
nintendo = company == 'Nintendo'
mojang = company == 'Mojang Studios'
rockstar = company == 'Rockstar Games'
ea = company == 'Electronic Arts'
pubg = company == 'PUBG Corporation'
namco = company == 'Namco'
relogic = company == 'Re-Logic / 505 Games'
bethesda = company == 'Bethesda Softworks'
blizzard = company == 'Blizzard Entertainment'
cd_projekt_red = company == 'CD Projekt'
xbox = company == 'Xbox Game Studios'
sega = company == 'Sega'
twok = company == '2K Games'
psygnosis = company == 'Psygnosis'
sony = company == 'Sony Computer Entertainment'
nintendo_pokemon = company == 'Nintendo / The Pokémon Company'

nintendo.sum(), mojang.sum(), rockstar.sum(), ea.sum(), pubg.sum(), namco.sum(), relogic.sum(), bethesda.sum(), blizzard.sum(), cd_projekt_red.sum(), xbox.sum(), sega.sum(), twok.sum(), psygnosis.sum(), sony.sum(), nintendo_pokemon.sum()

# Mojang, EA, PUBG, Namco, Re-Logic / 505 Games, Bethesda Softworks, Blizzard Entertainment, CD Projekt, Sega, Psygnosis, and 2K Games has made the least amount of games.
```

```python
"6.) Is being muliplatform compatible that impactful in terms of sales?"

plt.pyplot.figure(figsize=(12,8))
sns.boxplot(y = videogame_sales_data['Platform(s)'], x = videogame_sales_data['Sales'])
```

Being multiplatform doesn't 100% lead to increase in sales, as there are many games which are stuck to one platform that averages more sales than being multiplatform. Although multiplatform games has achieved record breaking sales, it doesn't lead to increase sales.

```python
"9.) Which company is least likely to make the next big game?"
temp_sales = sales_dict
least_sales = []
for i in range(3):
    min(temp_sales, key = temp_sales.get)
    least_sales.append(min(temp_sales, key = temp_sales.get))
    del temp_sales[least_sales[i]]
    
least_sales
```

Psygnosis is least likely to make the next big game because as doing research into their company, most of their games have been unable to gain the amount of sales as their biggest game, while companies like 2k have still been able to create games and be successful and Xbox is owned by Microsoft, whom has bought out Activision Blizzard King.

```python
"10.) Which company would be most likely to make the next big game?"

temp_sales = sales_dict
most_sales = []
for i in range(3):
    max(temp_sales, key = temp_sales.get)
    most_sales.append(max(temp_sales, key = temp_sales.get))
    del temp_sales[most_sales[i]]
    
most_sales
```

Nintendo is most likely to create the next big game. Nintendo has had a history of making one of the most popular games in history, having made one of the most sold games in the world, and most of their games are within the top 20 in this data set. Even with Activision having more resorces, Nintendo is most likely to create the next big game.
